```python
from tkinter import *
from tkinter import ttk


class Window:
    def __init__(self, master):
        def hide_me(event):
            self.pressed.eingabefeld.pack_forget()

        def selected(event):

            if self.clicked.get() == 'Keine Angabe':
                self.myLabel = Label(master, fg='#e63946',
                                     text="Hey, " + self.eingabefeld_wert.get() + " we accept that you do not want to provide information about your situation, but for the best experience we highly recommend it.").pack()

            else:
                pass

        self.Abschluss = [
            "Please select",
            "Student",
            "employee",
            "Master",
            "Keine Angabe",
        ]

        Label(master, text="Name").pack()

        self.eingabefeld_wert = StringVar()
        self.eingabefeld = Entry(master, textvariable=self.eingabefeld_wert)
        self.eingabefeld.pack()

        Label(master, text="Email").pack()

        self.eingabefeld_wert2 = StringVar()
        self.eingabefeld2 = Entry(master, textvariable=self.eingabefeld_wert2)
        self.eingabefeld2.pack()

        Label(master, text="Nationality").pack()

        self.eingabefeld_wert3 = StringVar()
        self.eingabefeld3 = Entry(master, textvariable=self.eingabefeld_wert3)
        self.eingabefeld3.pack()

        self.clicked = StringVar()
        self.clicked.set(self.Abschluss[0])

        self.drop1 = OptionMenu(master, self.clicked, *self.Abschluss, command=selected)
        self.drop1.pack(pady=20)

        def pressed():

            self.new_window = Toplevel(master)

            # sets the title of the
            # Toplevel widget
            self.new_window.title("Expensetracker")

            # sets the geometry of toplevel
            self.new_window.geometry("1080x800")

            self.myLabel_new = Label(self.new_window, fg='medium sea green',
                                     text="Hey, " + self.eingabefeld_wert.get() + " Start tracking your expenses by clicking on submit!").pack()

            # add some style
            style = ttk.Style()
            # pick a theme

            # configure our treeview colors
            # clam/ default/ alt/ = Ã¤ndert oberste reihe
            style.theme_use("clam")
            style.configure("Treeview",
                            background="white",
                            foreground="black",
                            rowheight=35,  # each row in treeview-> how fat or skinny
                            fieldbackground="white"
                            )
            # change selected color (wenn man drauf clickt die farbe)
            style.map('Treeview',
                      background=[('selected', 'green')])

            self.my_tree = ttk.Treeview(self.new_window)

            # Columns
            self.my_tree['columns'] = ("Product name", "Category", "Date")
            self.my_tree.column("#0", width=120, minwidth=30)
            self.my_tree.column("Product name", anchor=W, width=140)
            self.my_tree.column("Category", anchor=CENTER, width=100)
            self.my_tree.column("Date", anchor=W, width=140)

            # Headings
            self.my_tree.heading("#0", text="Label", anchor=W)
            self.my_tree.heading("Product name", text="Product name", anchor=W)
            self.my_tree.heading("Category", text="Category", anchor=CENTER)
            self.my_tree.heading("Date", text="Date", anchor=W)

            self.data = [
                ["Macbook Air", "Electronics", "21.06"],
                ["Macbook Pro", "Electronics", "21.06"],
                ["Acer Swift 3", "Electronics", "21.06"]
            ]
            global count
            count = 0
            for record in self.data:
                self.my_tree.insert(parent='', index='end', iid=count, text="parent",
                                    values=(record[0], record[1], record[2]))
                count += 1

            self.my_tree.pack(pady=20)

            self.add_frame = Frame(self.new_window)
            self.add_frame.pack(pady=20)

            self.pl = Label(self.add_frame, text="Product name")
            self.pl.grid(row=0, column=0)

            self.cl = Label(self.add_frame, text="Category")
            self.cl.grid(row=0, column=1)

            self.dl = Label(self.add_frame, text="Date")
            self.dl.grid(row=0, column=2)

            self.product_box = Entry(self.add_frame)
            self.product_box.grid(row=1, column=0)

            self.category_box = Entry(self.add_frame)
            self.category_box.grid(row=1, column=1)

            self.date_box = Entry(self.add_frame)
            self.date_box.grid(row=1, column=2)

            # Add an expense
            def add_expenses():
                global count
                self.my_tree.insert(parent='', index='end', iid=count, text="parent",
                                    values=(self.product_box.get(), self.category_box.get(), self.date_box.get()))
                count += 1

                self.product_box.delete(0, END)
                self.category_box.delete(0, END)
                self.date_box.delete(0, END)

            # Remove all the expenses
            def remove_all():
                for record in self.my_tree.get_children():
                    self.my_tree.delete(record)

            # Remove one specific expense
            def remove_one():
                delete_selection = self.my_tree.selection()[0]
                self.my_tree.delete(delete_selection)

            # Buttons
            self.add_expenses = Button(self.new_window, text="Add your Expense", command=add_expenses)
            self.add_expenses.pack(pady=20)

            # remove all
            self.remove_all = Button(self.new_window, text="Remove all Expenses", command=remove_all)
            self.remove_all.pack(pady=10)

            # remove one
            self.remove_one = Button(self.new_window, text="Remove a specific expense", command=remove_one)
            self.remove_one.pack(pady=20)
            pass

        self.button = Button(master, text='Submit', command=pressed, bg='medium sea green', fg='white')
        self.button.pack(pady=10)


root = Tk()
root.geometry("800x400")
my_gui = Window(root)
root.mainloop()

root.mainloop()
```
